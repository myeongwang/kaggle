```python
# missing value를 handling합니다.
# column을 지울까요 / 채울까요?
# drop할 columns들을 골라보자 -> 생존율이랑 관계 없을 것 같은거
# PassengerID, Name, Ticket, Cabin,

train= train.drop(columns=["PassengerId","Name","Ticket","Cabin"])
train.info()
```

```python
# "Age" column 채우기

train = train.fillna(train.Age.mean())
train.info() # age non-null 확인 완료
```

## Feature Engineering

### 1. Categorical feature encoding
### 2. Normalization

```python
# categorical feature --> one-hot Encoding, Ordinal Encoding

#1. ordinal Encoding --> ordinal feature를 변환할 때 쓰임. e.g. 학력, 선호도
#2. one-hot Encoding --> nominalfeature를 변환할 때 쓰임. e.g.성별,부서,출신학교...

pd.get_dummies(train, columns=["Sex","Embarked"]) #원핫인코딩, sex/Embarked로
# female 0, male 1/  CQS 중 탑승한 곳은 1, 아닌 곳은 0
```

```python
train_OHE= pd.get_dummies(train, columns=["Sex","Embarked"])
train_OHE
```

```python
# Normalization --> Min-Max scaling
# survived 0,1로 구분되어진 라벨링인데 이걸 정규화 하면 안되지 그래서 빼주자
X= train_OHE.drop(columns = "Survived") #input matrix
y= train_OHE.Survived #target vector
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# scaler.fit()
# scaler.transform()
#X_scaled =scaler.fit_transform(X) # 한 번에 가능
temp = scaler.fit_transform(X.loc[:,["Age","Fare"]])
X["Age"]=temp[:,0]
X["Fare"] = temp[:,1]
X
```

## Training

```python
# sklearn에서 배웠던 분류 모델들을 불러와 보자.
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

#평가지표
from sklearn.metrics import accuracy_score
```

```python
clf = SGDClassifier()
clf2 = LogisticRegression()
clf3 = DecisionTreeClassifier()
clf4= RandomForestClassifier()

clf.fit(X,y)
clf2.fit(X,y)
clf3.fit(X,y)
clf4.fit(X,y)

pred=clf.predict(X) # train에 대한 학습 결과
pred2=clf2.predict(X)
pred3=clf3.predict(X)
pred4=clf4.predict(X)
```

```python
print(" 1. Linear Classifier, Accuracy for training : %.4f" % accuracy_score(y,pred))
print(" 2. Logistic Regression, Accuracy for training : %.4f" % accuracy_score(y,pred2))
print(" 3. Decision Tree, Accuracy for training : %.4f" % accuracy_score(y,pred3))
print(" 4. RandomForest , Accuracy for training : %.4f" % accuracy_score(y,pred4))
```
